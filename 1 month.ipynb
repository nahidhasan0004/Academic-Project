```python

# Temperature data
temp_data = {
    "Year": [1981],
    "D_01": [23.9], "D_02": [24.4], "D_03": [24.6], "D_04": [25.6], "D_05": [27.2],
    "D_06": [26.6], "D_07": [21.1], "D_08": [20.6], "D_09": [22.2], "D_31": [24.4]
}

# Rainfall data
rainfall_data = {
    "Year": [1981],
    "D_01": [0], "D_02": [0], "D_03": [0], "D_04": [0], "D_05": [0],
    "D_06": [0], "D_07": [2], "D_08": [0], "D_09": [0], "D_31": [9]
}

# Create DataFrames
temp_df = pd.DataFrame(temp_data)
rainfall_df = pd.DataFrame(rainfall_data)

# Drop the 'Year' column for correlation calculation
temp_df_values = temp_df.drop(columns=['Year'])
rainfall_df_values = rainfall_df.drop(columns=['Year'])

# Calculate correlation between temperature and rainfall datasets
correlation = temp_df_values.iloc[0].corr(rainfall_df_values.iloc[0])

# Display the correlation
print("Correlation between temperature and rainfall:", correlation)
```
